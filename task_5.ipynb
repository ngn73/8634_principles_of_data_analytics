Load usual modules ...

```python
#import required modules 
import pandas as pd
import numpy as np
from sklearn import datasets as ds
import matplotlib.pyplot as plt

#load the Iris Dataset
ds_iris = ds.load_iris()

#Convert Iris Dataset to a numpy Array
arr_iris = np.array(ds_iris.data)
```

Use a Scatter Plot to compare the Iris features</br>
[0] sepal length (cm)</br>
[1] sepal width (cm)</br>
[2] petal length (cm)</br>
[3] petal width (cm)</br>


```python
#extract data arrays
sepal_length = arr_iris[:, 0]
sepal_width = arr_iris[:, 1]
petal_length = arr_iris[:, 2]
petal_width = arr_iris[:, 3]
```

***
##### Compare Sepal Width/Length with default Scatter plot

```python
fig, ax = plt.subplots()  # 'fig' is the figure, 'ax' is the axes

#plot scatter plot
ax.scatter(sepal_length, sepal_width)

#labels
ax.set_xlabel('sepal_length')
ax.set_ylabel('sepal_width')

plt.show()
```

Not much noticeable correlation

***
##### Compare Petal Width/Length with default Scatter plot

```python
fig, ax = plt.subplots()  # 'fig' is the figure, 'ax' is the axes

#plot scatter plot
ax.scatter(petal_length, petal_width)

#labels
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')

plt.show()
```

More noticeable Linear correlation

***
##### Compare Petal Width/Length with Scatter plot and color-coded species

We need to plot 3 separate plots for each species of Iris</br>
Use the array indexing and slicing syntax of **arr[y=n, 0]**

```python
fig, ax = plt.subplots()  # 'fig' is the figure, 'ax' is the axes

# Define the colors for the species
colors = ['red', 'green', 'blue']  # Color for Setosa, Versicolor, and Virginica

x_iris = ds_iris.data
y_iris = ds_iris.target
#plot scatter plot
# Plot each species with different colors
for i in range(3):
    ax.scatter(x_iris[y_iris == i, 2],  # Petal length (x-axis)
                x_iris[y_iris == i, 3],  # Petal width (y-axis)
                color=colors[i],  # Color for the species
                label=ds_iris.target_names[i],  # Label for the species
                alpha=0.7,  # Set transparency
                marker = '.',  # dot
                s=100)  # Size of the points

#labels
ax.set_xlabel('petal_length')
ax.set_ylabel('petal_width')

plt.show()
```

3 Species are clearly clustered in separate data groups 

***
### Try plot all combinations of Features for the Iris dataset in a 4 x 4 Grid

```python
# Define the colors for the 3 species
colors = ['red', 'green', 'blue']  # Color for Setosa, Versicolor, and Virginica

# Set up a 4x4 grid
fig, axes = plt.subplots(nrows=4, ncols=4, figsize=(15, 15))


feature_names = ds_iris.feature_names
x_iris = ds_iris.data
# Loop through each pair of features and plot them
for i in range(4):
    for j in range(4):
        ax = axes[i, j] #select specific axis for plot within 4x4 grid
        if i == j:
            # (similar feature names) ... Just print feature name
            ax.text(0.3, 0.5, feature_names[i])
        else:
            # (different feature pairs) ... plot a scatter plot
            for species in range(3):
                ax.scatter(x_iris[y_iris == species, i],  # feature i (x-axis)
                x_iris[y_iris == species, j],  # feature j (y-axis)
                color=colors[species],  # Color for the species
                label=ds_iris.target_names[species],  # Label for the species
                alpha=0.7,  # Set transparency
                marker = '.',  # dot
                s=150)  # Size of the points
            ax.set_xlabel(feature_names[i])
            ax.set_ylabel(feature_names[j])

plt.tight_layout()
plt.show()
```

***
#### Observations
* The most obvious linear correlation is in Petal length vs width
* The linear correlation in Sepal length vs width is not so obvious
* There is also an obvious linear correlation in Petal length vs Sepal length
* The 3 classes/species of Iris (Setosa, Versicolor, and Virginica) are clustered separately from one another (with the exception of Sepal width vs length that has overlapping data points )
